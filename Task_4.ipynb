```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/language_dataset.csv")

# Create a new DataFrame by stacking the language columns and creating labels
data_bulgarian = pd.DataFrame({'text': df['bulgarian'], 'language': 'bulgarian'})
data_english = pd.DataFrame({'text': df['english'], 'language': 'english'})

# Concatenate the two DataFrames
df_combined = pd.concat([data_bulgarian, data_english], ignore_index=True)

# Features & labels
X = df_combined['text']
y = df_combined['language']

# Convert text to numerical features
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_vec, y, test_size=0.2, random_state=42
)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
test_sentences = [
    "This is a good project",
    "தமிழ் என் தாய் மொழி",
    "आपका नाम क्या है",
    "Bonjour tout le monde"
]

test_vec = vectorizer.transform(test_sentences)
predictions = model.predict(test_vec)

for text, lang in zip(test_sentences, predictions):
    print(f"Text: {text} --> Language: {lang}")
```
