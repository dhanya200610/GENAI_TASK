```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report
```

```python
data = pd.read_csv("/content/Student_Performance.csv")

X = data[['study_hours', 'attendance_percentage', 'overall_score']]
y_reg = data['overall_score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

scaler_reg = StandardScaler()
X_train_reg = scaler_reg.fit_transform(X_train)
X_test_reg = scaler_reg.transform(X_test)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train_reg, y_train)

y_pred_reg = linear_model.predict(X_test_reg)

print("===== Linear Regression =====")
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_reg))
print("R2 Score:", r2_score(y_test, y_pred_reg))
```

```python
y_clf = data['final_grade']
X_train, X_test, y_train, y_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)

scaler_clf = StandardScaler()
X_train_clf = scaler_clf.fit_transform(X_train)
X_test_clf = scaler_clf.transform(X_test)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train_clf, y_train)

y_pred_clf = logistic_model.predict(X_test_clf)

print("\n===== Logistic Regression =====")
print("Accuracy:", accuracy_score(y_test, y_pred_clf))
print(classification_report(y_test, y_pred_clf))

```
